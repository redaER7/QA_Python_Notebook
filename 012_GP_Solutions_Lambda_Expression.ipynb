## Exercice 1

Return the words from a list of words which start with a target letter.

```python
words = ['Learn','aragorn','Not A','dog','amy','hi','aRMADA','This A','aRON']
target_letter = 'a'
```

```python
list(filter(lambda word:word[0] == target_letter,words))
```

## Exercice 3

Let's consider data <code>d</code> of Cooper, Coffee, Platinium Prices for a specific

1. Get Cooper prices
2. Get Coffee prices less than 180
3. Get Dates where Platinum prices are less than 1450
4. Add the following Gold prices ( respect same data structure )

Gold : 1259.3, 1256.3, 1245.6, 1243.7, 1244.3

<div>

</div>

```python
d = {'index': ['2014-05-28',
  '2014-05-29',
  '2014-05-30',
  '2014-06-02',
  '2014-06-03'],
 'columns': ['Cooper', 'Platinium', 'Coffee'],
 'data': [[3.19, 1465.1, 176.15],
  [3.15, 1462.1, 181.95],
  [3.14, 1454.7, 177.5],
  [3.18, 1438.7, 172.35],
  [3.15, 1435.5, 171.15]]}
d
```

**1. Get Cooper prices**

```python
list(map(lambda x: x[0],d['data']))
```

**2. Get Coffee prices less than 180**

```python
list(filter(lambda y : y<180,list(map(lambda x: x[2],d['data']))))
```

**3. Get Dates where Platinum is less than 1450**

```python
d['index']
```

```python
platinium = list(map(lambda x: x[1],d['data']))
platinium
```

```python
indexes = list(filter(lambda x: x[1] < 1450, enumerate(platinium)))
indexes
```

```python
idx,pr = list(zip(*indexes))
idx
```

```python
[d.get('index')[i] for i in idx]
```

**4. Add Gold prices**

```python
Gold = [1259.3, 1256.3, 1245.6, 1243.7, 1244.3]
```

```python
d['data']
```

```python
list(map(lambda x, y,: x.append(y), d['data'],Gold))
```

```python
[y.append(x) for x,y in zip(Gold,d['data'])]
```
